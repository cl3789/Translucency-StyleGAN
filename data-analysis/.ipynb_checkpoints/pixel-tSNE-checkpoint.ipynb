```python
import torch
import torchvision
from torchvision import transforms

import os
import copy
import numpy as np
import pandas as pd
from PIL import Image
import glob
import pickle
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.manifold import TSNE, MDS, LocallyLinearEmbedding
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import cross_validate, KFold,StratifiedKFold, GridSearchCV,ParameterGrid,train_test_split
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, f1_score

```

```python
trans_for_conv = transforms.Compose([
            transforms.Resize(512),                 
            transforms.ToTensor(),
            transforms.Lambda(lambda tensor:min_max_norm(tensor, 0, 1)),
            ])

def min_max_norm(X, min_value, max_value):
    """Min-max normalization. Refer to https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html"""
    X_std = (X - X.min()) / (X.max() - X.min())
    X_scaled = X_std * (max_value - min_value) + min_value
    return X_scaled

def cal_norm(imgs):
    mean_imgs = np.mean(imgs)
    std_imgs = np.std(imgs)
    imgs_norm = (imgs - mean_imgs)/std_imgs

    return imgs_norm, mean_imgs, std_imgs

def get_imgs(img_path):
    ## Get a list of images from directory
    img_names = glob.glob(img_path)

    imgs = []

    for label in img_names:
        print(label)
        img = Image.open(label)
        img_norm = trans_for_conv(img)
        img_norm = np.array(img_norm).transpose(1, 2, 0)
        img_norm = img_norm.ravel()
#         img_norm = np.expand_dims(img_norm.ravel(), axis=0)
#         print(img_norm.shape)
        
#         img_norm, _, _ = cal_norm(img)
#         img_norm = np.expand_dims(img_norm.ravel(), axis=0)
#         print(img_norm.shape)
#         imgs = np.append(imgs, img_norm, axis = 0)
        imgs.append(img_norm)
        
    return  np.asarray(imgs), img_names
```

# Dimensionality reduction of features

```python
class dimension_reduction():
    def __init__(self, 
                 features, ## numpy array
                 preprocess = False):
        self.features = features 
        self.preprocess = preprocess
        print(self.features.shape)
        
    def preprocess_data(self):
        X = StandardScaler().fit_transform(self.features)
        return X
    
    def apply_tsne(self, n_comp, perplexity, rand_state=1):
        tsne = TSNE(n_components=n_comp, verbose=1, perplexity=perplexity, n_iter=300, init="random",
                    method='exact',random_state = rand_state)
        if self.preprocess == True:
            X_std = self.preprocess_data()
            X_tsne = tsne.fit_transform(X_std)
            return X_tsne
        else:
            return tsne.fit_transform(self.features)
    
    def apply_mds(self, n_comp, rand_state=1):
        mds = MDS(n_components=n_comp, max_iter=300, random_state = 1, n_jobs = -1)
        if self.preprocess == True:
            X_std = self.preprocess_data()
            X_mds = mds.fit_transform(X_std)
            return X_mds
        else:
            return mds.fit_transform(self.features)
        
    def apply_lle(self, n_comp, n_neighbor, rand_state=1):
        lle = LocallyLinearEmbedding(n_components=n_comp, n_neighbors=n_neighbor, 
                                     max_iter=300, random_state = 1, n_jobs = -1)
        if self.preprocess == True:
            X_std = self.preprocess_data()
            X_lle = lle.fit_transform(X_std)
            return X_lle
        else:
            return lle.fit_transform(self.features)
    
```

```python
def cross_val_svm(all_features, all_labels):
    
    #Train test split
    (data_train,
     data_test,
     labels_train,
     labels_test,
     ) = train_test_split(all_features, all_labels, stratify=all_labels, test_size=0.30, random_state=2)

  
    inner_cv = StratifiedKFold(n_splits = 3, shuffle = True, random_state = 1)
    outer_cv = StratifiedKFold(n_splits = 3, shuffle = True, random_state = 1)

    # SVC hyperparams to optimize
    C = np.power(10., np.arange(-4, -1))

    # set up parameter grid
    params = {
      'classify__C': C
    }


    pipe = Pipeline([
      ('scale', StandardScaler()),
      ('classify', SVC(random_state = 1))
    ])

    grid_SVC = GridSearchCV(pipe, params, cv = inner_cv, refit=True)
  

    # Nested CV scores
    scores = cross_validate(grid_SVC,
                          X = data_train,
                          y = labels_train,
                          cv = outer_cv,
                          scoring = ['roc_auc', 'accuracy'],
                          return_estimator = True)

    auc = scores['test_roc_auc']
    accuracy = scores['test_accuracy']
    estimators = scores['estimator']

    best_params = []
    for i in estimators:
        best_param = i.best_params_
        best_params.append(best_param)
        print(best_param)
        print('\n')
    print("accuracy:",accuracy)
    print("mean accuracy:", accuracy.mean())
    print("std accuracy:", np.std(accuracy))
    print('\n')
  
    best_c_index = np.argmax(accuracy)
    print("best_c_index:", best_c_index)

    best_C = best_params[best_c_index]['classify__C']
    print("Best C and best kernel:", best_C)
    best_svc_model = SVC(C = best_C, kernel = 'linear')
    classifier = best_svc_model.fit(all_features, all_labels)

    train_acc = accuracy_score(y_true=labels_train, y_pred=classifier.predict(data_train))
    print("Train accuracy on all images: ", train_acc)


    a = classifier.coef_.reshape(1, 512).astype(np.float32)
    direction = a / np.linalg.norm(a)

    return classifier, direction
```

```python
def train_svm(all_features, all_labels, best_C=0.001):
    
    all_features = StandardScaler().fit_transform(all_features)
    classifier = SVC(C = best_C, kernel = 'linear')
    classifier.fit(all_features, all_labels)

    train_acc = accuracy_score(y_true=all_labels, y_pred=classifier.predict(all_features))
    print("Train accuracy on all images: ", train_acc)


    a = classifier.coef_.reshape(1, 512).astype(np.float32)
    direction = a / np.linalg.norm(a)

    return classifier, direction

def get_svm_prediction(train_feature, train_label, test_features, img_names, dim_reduc_method):
    
    #model, direction = cross_val_svm(train_feature, train_label)
    model, direction = train_svm(train_feature, train_label)
    distance_from_bound = model.decision_function(test_features)
    w_norm = np.linalg.norm(model.coef_)
  
    distance_from_bound = distance_from_bound/w_norm
    predicted_test_label = model.predict(test_features)

    layer_predict_test = pd.DataFrame({'image_name':img_names,
                    'predicted':predicted_test_label,
                    'distance_bound':distance_from_bound})
    save_name = "svm_prediction_"+ str(dim_reduc_method) +".csv"
    save_path = os.path.join("/Users/chenxiliao/Dropbox/Projects/StyleGAN2-ada Translucency/for-github/translucency-stylegan/data-analysis/data/svm_dim_reduction", save_name)
    layer_predict_test.to_csv(save_path)
```
